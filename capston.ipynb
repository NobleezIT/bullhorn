```python
# Melt the confirmed cases data to long format
confirmed_long = cases.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                                var_name='Date', value_name='Confirmed Cases')
confirmed_long['Date'] = pd.to_datetime(confirmed_long['Date'])
```

```python
confirmed_long.tail(10)
```

```python
deaths_long = deaths.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'],
                          var_name='Date', value_name='Death Cases')
# deaths_long['Date'] = pd.to_datetime(deaths_long['Date'])
```

```python
deaths_long['Date'] = pd.to_datetime(deaths_long['Date'])
```

```python
recovered_long = recovered.melt(id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long'],
                                var_name='Date', value_name='Recovered Cases')
# recovered_long['Date'] = pd.to_datetime(recovered['Date'])
```

```python
recovered_long['Date'] = pd.to_datetime(recovered_long['Date'])
```

```python
recovered_long
```

```python
# Merge
first_merge = confirmed_long.merge(deaths_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
```

```python
first_merge
```

```python
full_data = first_merge.merge(recovered_long, on=['Province/State', 'Country/Region', 'Lat', 'Long', 'Date'])
full_data
```

```python
full_data = full_data.drop(['Lat', 'Long'], axis=1)
```

```python
early_covid = full_data[:2021-12-31]
china_early = early_covid[early_covid['Country/Region'] == 'China']
```

```python
full_data[(full_data['Date'] == '2020-12-16') & (full_data['Country/Region'] == 'China')] 
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(full_data['Date'], full_data['Death Cases'])
plt.show()
```

```python
naija = full_data[full_data['Country/Region'] == "Nigeria"]
```

```python
naija = naija.set_index('Date')
```

```python

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(naija.index, naija['Death Cases'], label='Death Cases')
plt.xlabel('Date')
plt.ylabel('Number of Death Cases')
plt.title('COVID-19 Death Cases Over Time')
plt.legend()
plt.grid(True)
plt.savefig('newimage.png')
plt.show()
```
